```python
from qiskit.utils import QuantumInstance
from qiskit import IBMQ, Aer
from main import *
from QND_Detector_Tomography import Plot_Choi_Matrices, Fidelity
import QuantumTomography as qt
from joblib import Parallel, delayed
%precision 2
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q-csic', group='internal', project='iff-csic')
backend  = provider.get_backend('ibmq_casablanca')
noise_model = NoiseModel.from_backend(backend)
quantum_instance = QuantumInstance( backend, shots=2**13 )
```

## Circuits

```python
circ_single = measurement_process_tomography_circuits(7)
circ_double = measurement_process_tomography_circuits_v2(2,2)
```

```python
circ_gst = tomographic_gate_set_tomography_circuits(7)
```

```python
circ_0 = circ_single
```

```python
qubits_1 = [0,1,3,5]
circ_1 = []
for circ_loop in circ_double:
    circ = QuantumCircuit(7,8)
    circ.compose(circ_loop, qubits=qubits_1, inplace=True)
    circ_1.append( circ )
```

```python
qubits_2 = [1,3,5,6]
circ_2 = []
for circ_loop in circ_double:
    circ = QuantumCircuit(7,8)
    circ.compose(circ_loop, qubits=qubits_2, inplace=True)
    circ_2.append( circ )
```

```python
qubits_3 = [1,2,4,5]
circ_3 = []
for circ_loop in circ_double:
    circ = QuantumCircuit(7,8)
    circ.compose(circ_loop, qubits=qubits_3, inplace=True)
    circ_3.append( circ )
```

## Running the circuits

```python
circ_run_1 = circ_gst + circ_0 + circ_1
circ_run_2 = circ_2 + circ_3
```

```python
results = Parallel(n_jobs=-1)( delayed(quantum_instance.execute )(circ) for circ in [circ_run_1,circ_run_2] ) 
results_1 = results[0]
results_2 = results[1]
```

## Gate Set Tomography (GST)

```python
gateset = tomographic_gate_set_tomography_fitter( results_1, circ_gst ).fit()
```

```python
gateset
```

```python
states_gst= []
measures_gst = []
for m in range(7):
    rho = gateset[0][m]
    Pi  = gateset[1][m]
    Y   = gateset[2][m]
    states_gst_temp   = []
    measures_gst_temp = []
    for v in [ np.eye(4), Y[2], Y[3] ]:
        for u in [ np.eye(4), Y[1] ]:
            states_gst_temp.append( v@u@rho )
        measures_gst_temp.append( v.T.conj()@Pi )    
            
    states_gst.append( np.array(states_gst_temp).T )
    measures_gst.append( np.array(measures_gst_temp).transpose(1,0,2).reshape(4,-1)/3 )
    
states_gst   = np.array( states_gst )
measures_gst = np.array( measures_gst )
```
