```python
from qiskit.utils import QuantumInstance
from qiskit import IBMQ, Aer
from main import *
import QuantumTomography as qt
from joblib import Parallel, delayed
import json
from time import time
```

```python
datos = np.load('datos_montecarlo_2.npy',allow_pickle=True)
```

```python
q1s = []
q2s = []
q3s = []
for d in datos:
    q1s.append( d[0] )
    q2s.append( d[1] )
    q3s.append( d[2] ) 
```

```python
q1_mean = np.mean( q1s, axis=0 )
q1_std  = np.std( q1s, axis=0 )
q2_mean = np.mean( q2s, axis=0 )
q2_std  = np.std( q2s, axis=0 )
q3_mean = np.mean( q3s, axis=0 )
q3_std  = np.std( q3s, axis=0 )
```

```python
for k in range(3):
    plt.errorbar( x=range(7), y=q1_mean[:,k], yerr=q1_std[:,k], capsize=5, fmt='--o' )
plt.legend( ['F','Q', 'D'] )
```

```python
for k in range(3):
    plt.errorbar( x=range(6), y=q2_mean[:,k], yerr=q2_std[:,k], capsize=5, fmt='--o' )
plt.legend( ['F','Q', 'D'] )
plt.xticks( [0,1,2,3,4,5], [(1,3),(4,5),(3,5),(0,1),(1,2),(5,6)] );
```

```python
[[(1, 3), (4, 5)], [(3, 5), (0, 1)], [(1, 2), (5, 6)]]
```

```python
for k in range(2):
    plt.errorbar( x=range(6), y=q3_mean[:,k], yerr=q3_std[:,k], capsize=5, fmt='--o' )
plt.legend( ['cF','cQ'] )
plt.xticks( [0,1,2,3,4,5], [(1,3),(4,5),(3,5),(0,1),(1,2),(5,6)] );
```

```python
T = np.zeros((6,3))
T[0,:] = q2_mean[0,:] - q1_mean[1,:]*q1_mean[3,:] 
T[1,:] = q2_mean[1,:] - q1_mean[4,:]*q1_mean[5,:] 
T[2,:] = q2_mean[2,:] - q1_mean[3,:]*q1_mean[5,:] 
T[3,:] = q2_mean[3,:] - q1_mean[0,:]*q1_mean[1,:] 
T[4,:] = q2_mean[4,:] - q1_mean[1,:]*q1_mean[2,:] 
T[5,:] = q2_mean[5,:] - q1_mean[5,:]*q1_mean[6,:] 
T = abs(T)
```

```python
for k in range(2):
    plt.plot( 1-np.sqrt(T[:,k]), '--o' )
plt.legend( ['F','Q'] )
plt.xticks( [0,1,2,3,4,5], [(1,3),(4,5),(3,5),(0,1),(1,2),(5,6)] );
```
